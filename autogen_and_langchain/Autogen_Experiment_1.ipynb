```python
# Initial message from Customer Proxy to Project Manager
customer_message = (
    "The system should be capable of detecting whether a person is wearing a face mask with high accuracy."
    " I want the system to track performance over time and provide detailed logs of detection events."
    " Additionally, a simple web interface is required for users to interact with the system, view data, and manage settings."
)

# Project Manager's prompt to Scrum Master
project_manager_to_scrum_master_prompt = (
    "We have received the customer's requirements for a face mask detection system. "
    "Please organize the tasks in sprints and ensure the team is aligned with the project goals. "
    "Estimate the number of tasks required for sprint planning. "
    "Calculate the effort for each sprint. "
    "Keep track of progress and report any blockers."
)

# Project Manager's prompt to Requirement Engineer
project_manager_to_requirement_engineer_prompt = (
    "We have received the customer's request for a mask detection system."
    "Please identify and document all use cases and requirements for this project. "
    "Ensure the requirements are clear, detailed, and numbered (e.g., REQ-01, REQ-02). "
    "Provide at least 10 use cases, and estimate the time and effort needed to complete the requirement gathering process." 
    "Let me know once you’ve finished so we can move forward."
)
```

```python
# Initialize GroupChat
groupchat = autogen.GroupChat(
    agents=[
        customer_proxy_agent,
        project_manager_agent,
        scrum_master_agent,
        tester_agent,
        ui_ux_designer_agent,
        business_analyst_agent,
        devops_engineer_agent,
        security_engineer_agent,
        technical_writer_agent,
        requirement_engineer_agent,
        system_engineer_agent,
        software_engineer_agent,
        documentation_engineer_agent,
        my_agent,
    ],
    messages=[],
    speaker_selection_method="round_robin",  # Automatically selects the next speaker
    max_round=20,
)
```

```python
manager = autogen.GroupChatManager(
    groupchat=groupchat, llm_config=llm_config
)
```

```python
# Customer Proxy sends the initial message to the Project Manager
groupchat_result = customer_proxy_agent.initiate_chat(
    recipient=project_manager_agent,
    message=customer_message,
)

# Project Manager sends the prompt to the Requirement Engineer
groupchat_result = manager.initiate_chat(
    agent=project_manager_agent,
    recipient=scrum_master_agent,
    message=project_manager_to_scrum_master_prompt,
)
```
